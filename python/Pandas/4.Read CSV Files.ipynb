```python
import pandas as pd 
def main():
    try:
        df = pd.read_csv('dta.csv', header = None)

        print(df.to_string())
    except FileNotFoundError:
        raise f'Not found'
if __name__ == '__main__':
    main()
```

```python
import pandas as pd 
import os
def main():
    if(os.path.exists('data.csv')):
        df = pd.read_csv('data.csv', header = None)

        print(df.to_string())
    else:
        raise f'Not found'
if __name__ == '__main__':
    main()
```

```python
# max_rows
    # - The number of rows returned is defined 
    # - "pd.options.display.max_rows"
print(pd.options.display.max_rows)
print(df)

print(pd.options.display.max_rows)
```

```python
# Increase the maximum number of rows to display the entire DataFrame:

pd.options.display.max_rows = 9999
df = pd.read_csv('data.csv', header = None)
print(df)
```
